```python
it=gr.get_group('IT')
it['SAL'].sum()
```

**concatination**

```python
dummy1=pd.DataFrame({'ename':['A','B','C'],
                    'deptname':['IT','SALES','MARKETING'],
                    'deptno':[1,2,1]})
dummy2=pd.DataFrame({'deptno':[1,2,3],
                    'sal':[30000,25000,20000]})
```

```python
pd.concat([dummy1,dummy2])
```

```python
pd.concat([dummy1,dummy2],axis=1)
```

**joins**

```python
pd .merge(dummy1,dummy2,on='deptno',how='inner')
```

```python
pd .merge(dummy1,dummy2,on='deptno',how='right')
```

```python
pd .merge(dummy1,dummy2,on='deptno',how='left')
```
