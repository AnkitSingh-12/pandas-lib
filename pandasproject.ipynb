```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/titanic.csv')
df
```

**1. Write a query to retrieve all passengers who embarked from Southampton (Embarked =
'S').
**

```python
df=df[df['embarked']=='S']
df
```

**2. Filter the dataset to find passengers who paid a fare between 20 and 40 units of currency.**

```python
df=df[(df['fare']>=20) & (df['fare']<=40)]
df
#
```

**3. Filter the dataset to find passengers whose age is missing (Age = NaN).
**

```python
df=df[df['age'].isnull()]
df
```

**4. Write a query to retrieve passengers who had siblings or spouses aboard the Titanic
(SibSp > 0)**

```python
df=df[df['sibsp']>0]
df
```

**5. Write a query to find passengers who survived and paid more than the median fare.**

```python
df=df[df['survived']==1]
df=df[df['fare']>df['fare'].median()]
print(df)
```

**6. Write a query to retrieve passengers who were under the age of 18.**

```python
df=df[df['age']<18]
print(df)
```

**7. Write a query to retrieve only the passengers who were both female and survived**

```python
df=df[(df['sex']=='female') & (df['survived']==1)]
print(df)
```

```python
d = df[(df["survived"] == 1) & (df["sex"] == "female")]
print(d)
```

**8. Filter the dataset to retrieve passengers who traveled in either first or second class**

```python
df=df[(df['pclass']==1) | (df['pclass']==2)]
print(df)

```

**9. Filter the dataset to find children (passengers under 18) who survived the disaster.**

```python
df=df[(df['age']<18) & (df['survived']==1)]
print(df)

```

**10.Filter the dataset to find all female passengers who traveled in first class**


```python
df=df[(df['sex']=='female') & (df['pclass']==1)]
print(df)
```

```python
d= df[(df["sex"] == "female") & (df["pclass"] == 1)]
print(d)
```
