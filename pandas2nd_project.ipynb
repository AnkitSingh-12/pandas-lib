```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/titanic.csv')
df
```

**4)how many ppl survived in b5 cabin**

```python
sv= df[(df['cabin'] == "B5") & (df["survived"] == 1)]

count = len(sv)
print(f"Number of people who survived in cabin B5: {count}")
```

**5)count of females whose age is less than 20 and survived**

```python
dfa = df[(df['sex'] == 'female') & (df['age'] < 20) & (df['survived'] == 1)]
count = len(dfa)
print(f"Number of females under 20 who survived: {count}")
```

**6)Total fare paid by embarked S females**

```python

dfb = df[(df['embarked'] == 'S') & (df['sex'] == 'female')]
fare = dfb['fare'].sum()
print(f"Total fare paid by females who embarked at S: {fare}")
```

**7)count of ppl in pclass 3 having only 1 person with them**




```python

dfc = df[(df['pclass'] == 3) & (df['sibsp'] == 0) & (df['parch'] == 0)]
count = len(dfc)
print(count)
```

**8)maximum fare paid in embarked Q**

```python

dfd = df[(df['embarked'] == 'Q')]
max_fare = dfd['fare'].max()
print(max_fare)
```

```python
df1=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/hr_data.csv')
df1
```

**9) Avg time spend by the ppl in support department who got promoted**


```python
sf2 = df1[(df1['department'] == 'support') & (df1['promotion_last_5years'] == 1)]
avg_time_spent = sf2['time_spend_company'].mean()
print(avg_time_spent)
```

**10)avg number of projects ppl with high salary are handling in IT departmen**

```python
sf3 = df1[(df1['department'] == 'IT') & (df1['salary'] == 'high')]
avg_projects = sf3['number_project'].mean()
avg_projects
```

**11) Total number of projects handled by support department.**

```python
sf4 = df1[(df1['department'] == 'support')]
total_projects = sf4['number_project'].sum()
total_projects
```
